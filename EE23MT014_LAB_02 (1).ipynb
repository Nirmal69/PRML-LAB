1. **Write a python code to generate sample data from gaussian distribution with mean zero and covariance identity matrix $I$ i.e., $N(0,I)$.**

*   **Plot the histogram for the same.**
*   **Create a contour plot to visualize the generated data.**

```python
# Write your code here.
import numpy as np
import matplotlib.pyplot as plt

num_samples = 10000
mean = np.array([0, 0])
cov_mat = np.array([[1, 0],
                   [0, 1]])
data = np.random.multivariate_normal(mean, cov_mat, num_samples)

plt.subplot(2, 2, 1)
plt.hist(data[:, 0], bins=300, alpha=0.5, color='blue', label='Sample 1')
plt.hist(data[:, 1], bins=300, alpha=0.5, color='red', label='Sample 2')
plt.title('Histograms')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()

plt.subplot(2, 2, 2)
plt.hist(data[:, 0], bins=300, alpha=0.5, color='blue', label='Sample 1')

plt.title('Histograms Sample 1')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()

plt.subplot(2, 2, 3)

plt.hist(data[:, 1], bins=300, alpha=0.5, color='red', label='Sample 2')
plt.title('Histograms ample 2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()

plt.subplot(2, 2, 4)
plt.scatter(data[:, 0], data[:, 1], marker='.')
plt.title('Contour Plot')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

3. **Write a python code to generate sample data from two gaussian distributions with different mean and covariance matrix and show the contour plot for the same. $\mu_1=[0.2 , 0.8]^T, \mu_2=[0.9,0.1]^T$ and $C_1=\begin{bmatrix} 0.3 & 0.8\\ 0.8 &1
\end{bmatrix}, C_2=\begin{bmatrix} 1 & -0.3\\ -0.3 &0.1
\end{bmatrix}$.**

```python
# Write your code here.

import numpy as np
import matplotlib.pyplot as plt

sample = 10000
mean_1 = np.array([0.2, 0.8])
mean_2 = np.array([0.9, 0.1])
cov_1 = np.array([[0.3, 0.8],
                  [0.8, 1]])
cov_2 = np.array([[1, -0.3],
                  [-0.3, 0.1]])

data_1 = np.random.multivariate_normal(mean_1, cov_1, sample)
data_2 = np.random.multivariate_normal(mean_2, cov_2, sample)

data_n = np.array(list(data_1)+list(data_2))


plt.subplot(2, 2, 1)
plt.hist(data_n[:, 0], bins=300, alpha=0.5, color='blue', label='Sample 1')
plt.title('Histograms Sample 1')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()

plt.subplot(2, 2, 2)
plt.hist(data_n[:, -1], bins=300, alpha=0.5, color='red', label='Sample 2')
plt.title('Histograms Sample 2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()

plt.subplot(2, 2, 3)
plt.hist(data_n[:, 0], bins=300, alpha=0.5, color='blue', label='Sample 1')
plt.hist(data_n[:, 1], bins=300, alpha=0.5, color='red', label='Sample 2')
plt.title('Histograms')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()

plt.subplot(2,2,4)
plt.scatter(data_1[:, 0], data_1[:, 1], marker='.', label='Distribution 1')
plt.scatter(data_2[:, 0], data_2[:, 1], marker='.', label='Distribution 2')
plt.title('Sample Data from Two Gaussian Distributions')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

4. Write a python code to generate data from two different gaussian distributions with

*   Mean $\mu_1=[0.2 , 0.8]^T$ and $C_1=\begin{bmatrix} 0.3 & 0.8\\ 0.8 &1 \end {bmatrix}$,     with probability 0.4
*   mean $\mu_2=[0.9 , 0.1]^T$ and $C_2=\begin{bmatrix} 1 & -0.3\\ -0.3 & 0.1 \end {bmatrix}$, with probability 0.6.

```python
# Write your code here

import numpy as np
import matplotlib.pyplot as plt
import random

sample = 10000
mean_1 = np.array([0.2, 0.8])
mean_2 = np.array([0.9, 0.1])
cov_1 = np.array([[0.3, 0.8],
                  [0.8, 1]])
cov_2 = np.array([[1, -0.3],
                  [-0.3, 0.1]])



data_1 = np.random.multivariate_normal(mean_1, cov_1, sample)
data_2 = np.random.multivariate_normal(mean_2, cov_2, sample)


generated_samples = []
for i in range(sample):
    choice = random.choices([data_1, data_2], weights=[0.4, 0.6])[0]
    generated_samples.append(random.choice(choice))



generated_samples = np.array(generated_samples)
print(generated_samples.shape)
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
plt.hist(generated_samples[:, 0], bins=300, color='blue', alpha=0.7)
plt.title('Histogram for Feature 1')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(generated_samples[:, 1], bins=300, color='red', alpha=0.7)
plt.title('Histogram for Feature 2')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.scatter(generated_samples[:, 0],generated_samples[:, 1],alpha=0.7, marker='.')
plt.title('Contour Plot of Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()

```

5. For the problem in 3, plot a line that roughly separates the two clusters.

```python
# Write your code here


import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


sample = 10000
mean_1 = np.array([0.2, 0.8])
mean_2 = np.array([0.9, 0.1])
cov_1 = np.array([[0.3, 0.8],
                  [0.8, 1]])
cov_2 = np.array([[1, -0.3],
                  [-0.3, 0.1]])



data_1 = np.random.multivariate_normal(mean_1, cov_1, sample)
data_2 = np.random.multivariate_normal(mean_2, cov_2, sample)


data_n = np.array(list(data_1)+list(data_2))


# Combining the sets
X = np.vstack((data_1, data_2))
y = np.hstack((np.zeros(sample), np.ones(sample)))
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k', marker='o')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100), np.linspace(ylim[0], ylim[1], 100))
Z = lda.predict(np.c_[xx.ravel(), yy.ravel()])
# Reshape the predictions to match the grid shape
Z = Z.reshape(xx.shape)
# Plot the decision boundary
plt.contour(xx, yy, Z, colors='k', levels=[0.5], linestyles='dotted')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear Discriminant Analysis')
plt.show()


plt.figure(figsize=(12, 5))


plt.scatter(data_n[:, 0],data_n[:, 1],alpha=0.7, marker='.')
plt.title('Contour Plot of Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
