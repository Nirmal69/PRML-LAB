```python
# write your code here.
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import time

img_path = 'prml_image.jpg'
img = Image.open(img_path)
plt.figure(figsize=(9,6))
plt.title('Original')
plt.imshow(img)


# CONVERT IMAGE TO GRAYSCALE

imggray= img.convert('LA')
plt.figure(figsize=(9,6))
plt.title('Gray')
plt.imshow(imggray)

# CONVERT IMG TO A MATRIX

imgmat= np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
# plt.figure(figsize = (9,6))
# plt.title('MATRIX VERSION')
# plt.imshow(imgmat, cmap = ('gray'))




#SVD

U, sigma, V = np.linalg.svd(imgmat)
print('shape of U :',U.shape, '\n', 'shape of V :', V.shape, '\n', 'shape of sigma :',  sigma.shape, '\n shape of imgmat', imgmat.shape )

# if i only keep the first n "term_number" principal components


for compression_ratio in range(10, 101, 10):
  num_singular_values = int((1-compression_ratio) * min(imgmat.shape)) # No.of singular values to keep
  reconstimg= np.matrix(U[:, :compression_ratio])*np.diag(sigma[:compression_ratio])*np.matrix(V[:compression_ratio, :])
  plt.figure(figsize=(9,6))
  Q = 100 - compression_ratio
  plt.title( str(Q) +'% COMPRESSED')
  plt.imshow(reconstimg, cmap='gray')


```

3. Consider the [breast cancer dataset](https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic). Write a python code to:

*   Use Principal Component Analysis (PCA) to reduce the 30 features in this dataset to lower dimension.
*   Then use the dimensionality reduced data to classify the tumor as malignant and benign.

```python
# Write your code here.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Load the breast cancer dataset
cancer = load_breast_cancer()

# Create DataFrame
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])

# Standardize the data
sc = StandardScaler()
scaled_data = sc.fit_transform(df)

# Define logistic regression function
def logistic(x_pca, n_components):
    x = x_pca
    y = cancer['target']

    # Split data into train and test sets
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

    # Train logistic regression classifier
    classifier = LogisticRegression(random_state=0)
    classifier.fit(x_train, y_train)

    # Predict
    y_pred = classifier.predict(x_test)

    # Print results
    print('RESULT FOR', n_components, 'FEATURES')
    # print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))

    # Calculate and print accuracy
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)

# Apply PCA and logistic regression for different number of components
for n_components in range(1, 30, 3):
    pca = PCA(n_components=n_components)
    x_pca = pca.fit_transform(scaled_data)

    print('Shape of original data:', scaled_data.shape)
    print('Shape after applying PCA to the data:', x_pca.shape)
    logistic(x_pca, n_components)
    print('----------------------------------------------------------------------------------------------------------------------------------------\n')
```
