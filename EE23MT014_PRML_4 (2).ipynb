## Without Using Inbuilt Functions

```python
#Write your code here.
import numpy as np
import matplotlib.pyplot as plt
data_file = open("SALES.txt","r")
X=[]
y=[]
for i in data_file.readlines()[:-1]:
    y.append(float(i.split()[1]))
    X.append(float(i.split()[0]))
data_file.close()
X = np.array(X)
y = np.array(y)
plt.scatter(X,y)
plt.title('Data points')
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20, random_state=42)

tot_samples = len(X_train)
x_mean = np.mean(X_train)
y_mean = np.mean(y_train)

numerator = 0
denominator = 0
for i in range(tot_samples):
    numerator += (X_train[i] - x_mean) * (y_train[i] - y_mean)
    denominator += (X_train[i] - x_mean) ** 2

b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean)
print("Slope (m) : ",b1)
print("Intercept (c) : ",b0)

#Testing
sq_er=0
for inp,out in zip(X_test,y_test):
    pred = b1 * inp + b0
    error = (out - pred)**2
    sq_er+=error
mean = sq_er/len(X_test)
rmse = mean**0.5
print("Mean Squared Error : ",mean)
print("RMSE Error:",rmse)


xx = np.linspace(min(min(X_train),min(X_test)),max(max(X_train),max(X_test)),tot_samples) ###plotting best line
yy = b1*xx + b0
plt.plot(xx,yy,color='red', label='Best Fit Line')
plt.scatter(X_train,y_train, color = 'g', label = 'Training Data')
plt.scatter(X_test,y_test, color = 'b', label = 'Training Data')
plt.title('Training,Testing and Fit Line')
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.show()
```

# 3.
**Classification Problem**. The dataset is the [Iris dataset](https://archive.ics.uci.edu/ml/datasets/iris). Here we have samples from 3 different flower species, and for each sample you have 4 different features that describe the flower. The task is to build a model which uses these features to predict the species of the flower. You can use the **K Nearest Neighbor (KNN)** algorithm. Demonstrate that you can achieve an accuracy of almost $100\%$ for this task using K-means algorithm. Comment on the difficult of the task by showing different images, and analyzing it.



### K Nearest Neighbor KNN

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

iris_data_file = open('iris.data','r')
lines = iris_data_file.readlines()[:-1]
iris_data = []
for i in lines:
    try:
        point = i[:-2].split(',')
        if point[-1] == 'Iris-setos':
            label = 0
        elif point[-1] == 'Iris-versicolo':
            label = 1
        else:
            label = 2

        a=float(point[0])
        b=float(point[1])
        c=float(point[2])
        d=float(point[3])
        one_point = [a,b,c,d,label]
        iris_data.append(one_point)
    except:
        continue

iris_data = np.array(iris_data)

X = iris_data[:,:-1]
y = iris_data[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20, random_state=42)
print('Training Data  : ',X_train.shape,y_train.shape)
print('Testing Data   : ',X_test.shape,y_test.shape)

classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
```

### K Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3,n_init='auto')
kmeans.fit(X_train)
y_predd = kmeans.predict(X_test)
accuracy = sum(y_predd == y_test) / len(y_predd)
print(accuracy)
```

# 4.
**Classification using Logistic Regression** [Breast Cancer dataset](https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic). This is another famous dataset with the interesting task to predict if a cancer cell is good or bad (or in medical terms: **malignant** or **benign**). In this dataset, you have 30 different features for each cancer cell that have been computed from medical images. Your task is to build a ML model and show that the model achieves atleast $95\%$ accuracy. Toward that, use the **Logistic Regression model**. This is similar to the Linear Regression model in the beginning.

```python
breast_datafile = open("wdbc.data",'r')
breast_datalines = breast_datafile.readlines()
breast_datafile.close()
```

```python
import numpy as np
breast_data_X=[]
breast_data_y=[]
for line in breast_datalines:
    one_data = line.replace('\n','')
    one_data = one_data.split(',')
    label = one_data.pop(1)
    one_data = [float(i) for i in one_data[1:]]
    breast_data_X.append(one_data)
    breast_data_y.append(label)
X = np.array(breast_data_X)
y = np.array(breast_data_y)
y = np.where(y == 'M', 1, 0)
print(X.shape,y.shape)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20, random_state=23)
print('Training Data  : ',X_train.shape,y_train.shape)
print('Testing Data   : ',X_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

# LogisticRegression
logi_regr = LogisticRegression(random_state=0,max_iter=10000)
logi_regr.fit(X_train, y_train)
y_pred = logi_regr.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy  :", accuracy*100)
print("Precision :", precision*100)
print("Recall    :", recall*100)
print("F1 Score  :", f1*100)
```
